# Environment class

```python
class environment:
    # States are represented as integers. Hashing the board state as a ternary number
    
    def __init__(self):
        self.state = 0
        self.board = np.zeros((3,3))
        self.num_states = 3**9        
    
    def draw_board(self):
        
    def game_over(self,force_recalculate=False):
        
    def get_state(self):
        # Converts board state to integer
        k = 0
        h = 0
        v = 0
        # Loop through all the board locations
        for i in range(3):
            for j in range(3):
                if self.board[i,j] == 0:
                    v = 0
                elif self.board[i,j] == self.x:
                    v = 1
                elif self.board[i,j] == self.o:
                    v = 2
                h += (3**k) * v
                k += 1
            #endfor
        #endfor
        return h
    
    def get_state_hash_and_winner(env,i=0,j=0):
        results = []
        
        for v in (0, env.x, env.o):
            env.board[i,j] = v # if empty board, it should already be 0
            if j == 2:
                # reset j back to 0, increasi i (unless i==2), then done
                if i == 1:
                    # the board is full, collect results and return
                    state == env.get_state()
                    ended = env.game_over(force_recalculate=True)
                    winner = env.winner
                    results.append((state,winner,ended))
                else:
                    results += get_state_hash_and_winner(env,i+1,0)
            else:
                # increment j, i stays the same
                results += get_state_hash_and_winner(env,i,j+1)
                
        return results
        
```

# Agent class

```python
class agent:
    
    def update_state_history(self, state):
        
    def update(self, env):
        
    def take_action(self, env):
        
    def initialV_x(env, state_winner_triples):
        # Initialises value function as follows
        # if x winds, V(s) = 1
        # if x loses or draw, V(s) = 0
        # otherwise, V(s) = 0
        V = np.zeros(env.num_states)
        for state, winner, ended in state_winner_triples:
            if ended:
                if winner == env.x:
                    v = 1
                else:
                    v = 0
            else:
                v = 0.5
            V[state] = v
        return V
    
    def initialV_o(env, state_winner_triples):
        # Initialises value function as follows
        # if o winds, V(s) = 1
        # if o loses or draw, V(s) = 0
        # otherwise, V(s) = 0
        V = np.zeros(env.num_states)
        for state, winner, ended in state_winner_triples:
            if ended:
                if winner == env.o:
                    v = 1
                else:
                    v = 0
            else:
                v = 0.5
            V[state] = v
        return V
```

# Main game loop
Takes in two players (agents) p1 and p2, and an environment env.

```python
class tictactoe:
    
    
    def __init__(self):
        current_player = None # The player that is currently active
    
    def play_game(p1, p2, env, draw=False):
        # Loops the game until its over
        current_player = None
        while not env.game_over():
            # Alternate between players, with p1 being first
            if current_player == p1:
                current_player = p2
            else:
                current_player = p1
                
            # Draw the board
            if draw:
                if draw == 1 and current_player == p1:
                    env.draw_board()
                if draw == 2 and current_player == p2:
                    env.draw_board()
                    
            # Get the current player to make a move
            current_player.take_action(env)
            
            # update state histories
            state = env.get_state()
            p1.update_state_history(state)
            p2.update_state_history(state)
        # end while    
        
        if draw:
            env.draw_board()
            
        # do value function updates
        p1.update(env)
        p2.update(env)
```
